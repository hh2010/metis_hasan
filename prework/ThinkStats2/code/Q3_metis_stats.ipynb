<b>The numbers generated by random.random are supposed to be uniform between 0 and 1; that is, every value in the range should have the same probability.

Generate 1000 numbers from random.random and plot their PMF and CDF. Is the distribution uniform?</b>

```python
# Import the modules needed to complete this exercise

from __future__ import print_function, division

%matplotlib inline

import numpy as np

import nsfg
import first
import thinkstats2
import thinkplot
import random
```

```python
# Create a list of 1000 random numbers from random.random() function

rand_list = [random.random() for x in range(1000)]
```

Since we are generating random floating point numbers and not rounding them, there are mostly unique values in this sample.  Therefore it is hard to tell if distribution is uniform by just looking at the plot.

```python
# Create the pmf and plot it
pmf = thinkstats2.Pmf(rand_list, label='random num pmf')

thinkplot.Hist(pmf, linewidth=.1)
thinkplot.Config(xlabel='random nums', ylabel='PMF')
```

```python
# Create the CDF and plot it

cdf = thinkstats2.Cdf(rand_list, label = 'random num cdf')

thinkplot.Cdf(cdf)
thinkplot.Show(xlabel='random nums', ylabel='CDF')
```

<b>Since the CDF plot approximates a straight line, we can describe the underlying data as being uniformly distributed.</b>
