For a space that is 'populated':

Each cell with one or no neighbors dies, as if by solitude.

Each cell with four or more neighbors dies, as if by overpopulation.

Each cell with two or three neighbors survives.

For a space that is 'empty' or 'unpopulated'

Each cell with three neighbors becomes populated.

```python
from IPython import display
display.clear_output(wait=True)

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
m = np.zeros([10,10])
m[4,4] = 1
m[6,4] = 1
m[6,3] = 1
m[6,5] = 1
m[5,5] = 1
m
```

```python
def check(m, i, j, newmap):
    neighbors = 0
    n = [(i,j+1),(i+1,j+1),(i+1,j),(i-1,j), 
         (i-1,j-1), (i-1,j+1), (i,j-1), (i+1,j-1)]
    for nbr in n:
        try:
            neighbors += int(m[nbr])
        except:
            pass

    if m[i, j] == 1:
        if neighbors <= 1:
            newmap[i, j] = 0
        elif neighbors >= 4:
            newmap[i, j] = 0
        else:
            newmap[i, j] = 1
    else:
        if neighbors == 3:
            newmap[i, j] = 1
            
def run(n):
    m = np.zeros([10,10])
    m[4,4] = 1
    m[6,4] = 1
    m[6,3] = 1
    m[6,5] = 1
    m[5,5] = 1
    print(m)
    plt.imshow(m)

    row, col = m.shape
    
    for k in range(n):
        newmap = np.zeros(m.shape)
        for i in range(row):
            for j in range(col):
                check(m, i, j, newmap)
        m = newmap
        print(m)
        plt.figure()
        plt.imshow(m)
```

```python
run(10)
```
